```python
import pandas as pd 
df = pd.read_csv("carbon_emission_dataset_with_Industry.csv") 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.dtypes)
```

```python
columns = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

for col in columns:
    print("\n", col)
    print("Mean =", df[col].mean())
    print("Median =", df[col].median())
    print("Min =", df[col].min())
    print("Max =", df[col].max())
```

```python
import matplotlib.pyplot as plt

df['Total_Energy_Consumption_kWh'].hist(bins=30)

plt.title('Total Energy Consumption Distribution')
plt.xlabel('Energy Consumption')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

columns = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

for col in columns:
    plt.figure(figsize=(6,4))
    df[col].hist(bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import matplotlib.pyplot as plt

columns = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

counts = [df[col].count() for col in columns]

plt.figure(figsize=(8,5))
plt.bar(columns, counts)
plt.title("Frequency of Variables")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['Total_Energy_Consumption_kWh'],
         label='Total Energy')

plt.plot(df['Date'], df['Renewable_Energy_Consumption_kWh'],
         label='Renewable Energy')

plt.plot(df['Date'], df['NonRenewable_Energy_Consumption_kWh'],
         label='Non Renewable Energy')

plt.legend()
plt.title("Energy Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Energy")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['float64','int64'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.boxplot(
    x='Industry_Sectors',
    y='Carbon_Emission_tCO2e_TARGET',
    data=df
)

plt.xticks(rotation=45)
plt.title('Carbon Emission by Industry Sector')

plt.show()
```

```python
selected = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Carbon_Emission_tCO2e_TARGET'
]

sns.pairplot(df[selected])

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(df['Date'], df['Carbon_Emission_tCO2e_TARGET'])

plt.title('Carbon Emissions Over Time')
plt.xlabel('Date')
plt.ylabel('Carbon Emission')

plt.show()
```

```python
df = df.sort_values('Date')

df['Moving_Average'] = (
    df['Carbon_Emission_tCO2e_TARGET']
    .rolling(window=30)
    .mean()
)

plt.figure(figsize=(12,5))

plt.plot(df['Date'],
         df['Carbon_Emission_tCO2e_TARGET'],
         alpha=0.4)

plt.plot(df['Date'],
         df['Moving_Average'],
         linewidth=2)

plt.title('30-Day Moving Average')
plt.show()
```

```python
df['Rolling_Variance'] = (
    df['Carbon_Emission_tCO2e_TARGET']
    .rolling(window=30)
    .var()
)

plt.figure(figsize=(12,5))

plt.plot(df['Date'],
         df['Rolling_Variance'])

plt.title('Rolling Variance')
plt.show()
```
